```python
cols_no_interest =['RowNumber', 'AnimalName', 'AnimalGender', 'AnimalBirthMonth','ZipCode','CommunityDistrict','NTA',
                    'CityCouncilDistrict','CongressionalDistrict', 'StateSenatorialDistrict','LicenseExpiredDate']
dogdf.drop(columns=cols_no_interest, inplace=True)
dogdf.info()
```

```python
manhattandog = dogdf[(dogdf['Borough']=='Manhattan')]
manhattandog.reset_index(drop=True, inplace=True)
manhattandog.info()
```

```python
dgcount = manhattandog.groupby(['BreedName'])[['Borough']].count()
dgcount.sort_values('Borough', ascending=False).head(5)
```

```python
manhpopdog = manhattandog[manhattandog['BreedName'].isin(['Yorkshire Terrier','Chihuahua','Shih Tzu'])]
manhpopdog.reset_index(drop=True, inplace=True)
manhpopdog.head(100)
```

```python
# We have two problems: the census tract information is in an inconsistent format; and, 'LicenseIssuedDate' should be in datetime format.
# We will fix them here:
manhpopdog['LicenseIssuedDate']= pd.to_datetime(manhpopdog['LicenseIssuedDate'])
manhpopdog.info()
```
